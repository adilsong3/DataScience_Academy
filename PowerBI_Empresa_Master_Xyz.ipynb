* **Empresa Master Xyz** e uma  rede  de  varejo fundada em 2020 no auge da pandemia que vende produtos eletrônicos e eletrodomésticos com lojas espalhadas por diversas cidades do Brasil. A mesma tem como objetivo no ano de 2023 crescer em 35% e para isso contratou uma equipe de dados para extrair os insights necessários para que o crescimento ocorra.

## **Estratégia de Vendas para o Próximo ano**

* A empresa está montando a estratégia de vendas para o próximo ano e precisa saber inicialmente saber qual dos fabricantes dos produtos vendidos, apresenta melhor desempenho nas vendas. O objetivo é descartar  os  fabricantes  cujos  produtos  possuem  poucas  vendas  e  tentar  negociar  melhores condições com os principais fabricantes.

## **Fonte de Dados**
* Um arquivo excell extraido pelo time de engenharia com dados coletados do sistema de vendas, CRM e ERP da empresa.
* Arquivo está dividido em 5 tabelas.

## **Dicionário de Dados**

* **ID-Produto >** Identificador único de cada produto
* **Produto >** Nome do produto
* **Categoria >** Categoria do produto
* **Segmento >** Segmento do produto
* **Fabricante >** Fabricante do produto
* **Loja >** Identificador único de cada loja
* **Cidade >** Cidade da loja onde foi efetuada a venda
* **Estado >** Estado da loja onde foi efetuada a venda
* **Vendedor >** Nome do vendedor
* **ID-Vendedor >** Identificador único de cada Vendedor
* **DataVenda >** Data da venda
* **ValorVenda >** Valor da venda

## **Divisão das tabelas**

* **Tabela produto (DIM_PRODUTO):**
     * ID-Produto (STRING) - PK
     * Categoria (STRING)
     * Fabricante (STRING)
     * Produto (STRING)
     * Segmento (STRING)

* **Tabela loja (DIM_LOJA):**
     * Loja (STRING) - PK
     * Cidade (STRING)
     * Estado (STRING)

* **Tabela vendedor (DIM_VENDEDOR):**
     * ID-Vendedor (INT) - PK
     * Vendedor (STRING)

* **Tabela tempo (DIM_TEMPO):**
     * Data Venda (DATE) - PK

* **Tabela de ligação (TB_FATO):**
     * ID-Produto (STRING) - PK | FK
     * ID-Vendedor (INT) - PK | FK
     * Loja (STRING) - PK | FK
     * Data Venda (DATE) - PK | FK
     * Valor Venda (Float)

***Legenda:***
* PK = Primary Key (Chave primária)
* FK = Foreign Key (Chave estrangeira)

## **Definindo Criando o Schema**
* Como os dados foram dividos em 4 tabelas dimensões e 1 tabela de ligação (fato), vou utilizar o Star Schema (Esquema no formato Estrela).
